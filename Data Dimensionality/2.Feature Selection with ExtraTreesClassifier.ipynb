### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.axes as ax
import seaborn as sns

from sklearn.ensemble import ExtraTreesClassifier
sns.set()
```

### Loading Data

### Cleaning Data

```python
## Hours where solar radiation is zero
clean1 = data[data['Hour']<6]
clean2 = data[data['Hour']>18]
```

```python
clean1.head()
```

```python
clean2.head()
```

```python
data = data.drop(clean1.index,axis=0)
data = data.drop(clean2.index,axis=0)
```

```python
data.head()
```

```python
data.info()
```

### Preparing input and output data

```python
x = data.iloc[:,0:10]
y = data.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
x = pd.DataFrame(MinMaxScaler().fit(x).transform(x))
x.head()
```

# <font color="green">Feature Selection with Extra Trees Classifier</font>

```python
model = ExtraTreesClassifier()
```

```python
model.fit(x,y)
```

```python
imp = pd.Series(model.feature_importances_,index=data.columns[0:-1])
imp
```

### Visualization

```python
imp.nlargest(5).plot(kind='bar')
```
